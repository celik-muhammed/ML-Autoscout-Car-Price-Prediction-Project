<p style="font-size:18px; font-weight:bold;">Some Reminders on Exploratory data analysis (EDA)</p> 

Exploratory data analysis (EDA) is an especially important activity in the routine of a data analyst or scientist. It enables an in depth understanding of the dataset, define or discard hypotheses and create predictive models on a solid basis. It uses data manipulation techniques and several statistical tools to describe and understand the relationship between variables and how these can impact business. By means of EDA, we can obtain meaningful insights that can impact analysis under the following questions (If a checklist is good enough for pilots to use every flight, it’s good enough for data scientists to use with every dataset).
1. What question are you trying to solve (or prove wrong)?
2. What kind of data do you have?
3. What’s missing from the data?
4. Where are the outliers?
5. How can you add, change or remove features to get more out of your data?

**``Exploratory data analysis (EDA)``** is often an **iterative brainstorming process** where you pose a question, review the data, and develop further questions to investigate before beginning model development work.
```python
def drop_high_null_percentage(df, percentage=95, inplace=True):
    for j, i in enumerate(df.columns):
        per = df[i].isna().sum() / df[i].isna().count()*100
        if per > percentage:
            print(j, 'Null Percentage:', per, i, 'dropping')
            df.drop(columns=i, inplace=inplace)
```

```python
def get_diff_category_column(cat_column, df=df, exclude=',/\n', pattern=r'[,\n]| /', strip='''\n' "!?|.,*+-_/]['''):
    """
    exclude  - satırlar da kategori ayırım yerleri
    pattern  - satırdaki veriyi ayırma regex
    strip    - alakasız işaretleri kaldırma
    """
    import re
    column = df[cat_column].dropna().apply(str).str.strip(strip)
    diff_value = list()
    for row in column:
        if not any(x in exclude for x in row) and row not in diff_value:
            diff_value.append(row)
        else:
            for data in map(lambda x: x.strip(strip), filter(None, re.split(pattern, row))):
                if data not in diff_value:
                    diff_value.append(data)
    return dict(enumerate(sorted(diff_value)))
```

```python
def get_regex_search_key(key_list):
    keys = '{}'.format(f'({"".join("{}|".format(w) for w in key_list)})').replace('|)', ')')
    return keys
```

```python
benzine  = ["Gasoline", "Regular", "Others", "Super"]
diesel   = ["Biodiesel", "Diesel"]
lpg      = ["Biogas", "CNG", "Domestic gas H", "LPG"]
hibrit   = ['Electric/Gasoline']
Electric = ['Electric']

def fueltype(row):
    if all(i in row for i in hibrit):
        return "Hybrit"
    elif any(i in row for i in Electric):
        return "Electric"
    elif any(i in row for i in lpg):
        return "LPG/CNG"
    elif any(i in row for i in benzine):
        if "Particulate Filter" in row :
            return "Benzine (Particulate Filter)"
        else :
            return "Benzine"
    elif any(i in row for i in diesel):
        if "Particulate Filter" in row :
            return "Diesel (Particulate Filter)"
        else :
            return "Diesel"
```

```python
def get_rows_diff_category(Series:pd.Series, exclude=''',/\n''', pattern=r'''[,\n]| /''', strip='''\\n' "!?|.,*+-_/]['''):
    import re
    df = pd.DataFrame()
    df[Series.name]  = Series.dropna().apply(str).str.strip(strip)
    df['category'] = df[Series.name].apply(lambda x: {*map(lambda x: x.strip(strip), filter(None, re.split(pattern, x)))})
    df['len'] = df['category'].apply(lambda x: len([*x]))
    df = df.sort_values(by='len', ascending=False)
    
    df_groups = {}
    def get_key_from_value(val, dic=df_groups):
        keys = [k for k, v in dic.items() for d in v if d == val]
        return keys[0] if keys else False
            
    for row in df['category']:  
        """
        scan top create columns uniquely
        """    
#         isalnum = all(x.isalnum() for x in row) 
        check_col = any(any(col in r for r in row) for col in df_groups.keys())
        check_col_r = any(any(r in col for r in row) for col in df_groups.keys())
        
        check_data = any(any(data in r for r in row) for data in {d for data in df_groups.values() for d in data})
        check_data_r = any(any(r in data for r in row) for data in {d for data in df_groups.values() for d in data})
        
        condition = not ((check_col or check_col_r) or (check_data or check_data_r)) 
        
        if len((*row,)) and condition:  
#             df_groups[next(iter(row))] = {*row}
            df_groups[[*row][0]] = {*row}
    
    for row in df['category']:   
        """
        scan top fill column 
        """
        for data in row:  
#             check_col   = any(col in data for col in df_groups.keys())
#             check_col_r = any(data in col for col in df_groups.keys())                
#             check_data   = any(d in data for d in {d for data in df_groups.values() for d in data})
#             check_data_r = any(data in d for d in {d for data in df_groups.values() for d in data})                
#             check_data   = any(any(d in data for d in setd) for setd in df_groups.values())
#             check_data_r = any(any(data in d for d in setd) for setd in df_groups.values())
        
#             get_col_r = {*filter(None, map(lambda k: k if data in k else False, filter(None, df_groups.keys())))}
#             get_val_r = {*filter(None, map(lambda d: data if data in d else False, filter(None, {d for data in df_groups.values() for d in data})))}
#             condk_r = get_val_r.pop() if any(get_val_r) else False
#             condc_r = get_val_r.pop() if any(get_val_r) else False

            get_col = {*filter(None, map(lambda k: k if k in data else False, filter(None, df_groups.keys())))}
            get_val = {*filter(None, map(lambda d: d if d in data else False, filter(None, 
                                                    {d for data in df_groups.values() for d in data})))}

            cond_col = get_col.pop() if any(get_col) else False
            cond_data = sorted(get_val, reverse=True).pop() if any(get_val) else False
            cond_key = get_key_from_value(cond_data, df_groups)
             
            if cond_col: 
                df_groups[cond_col] = df_groups[cond_col].union({(data)})

            if cond_key: 
                df_groups[cond_key] = df_groups[cond_key].union({(data)})
        
    for k,v in df_groups.items():
        df_groups[k]=[*v]
    return pd.DataFrame.from_dict(data=df_groups, orient='index').T
```
